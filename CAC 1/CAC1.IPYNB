```python
# Function defined to remove any punctuations and convert every word to lowercase for easy processing
def mine_word(word):
    return word.strip('.,!?').lower()

# Function defined to perfrom all the necessary segregation into postive and negative reviews
def feedback_analysis(feedback_data):

# Initializing counters for feedback analysis
    pos_count = 0
    neg_count = 0   
    feedback_data
# Initializing two lists to store all the 'Postive' and 'Negative' words used in the file respectively
    pos_list = []
    neg_list = []

    for feedback in feedback_data: # Iterating through the lines in the text file
        words = feedback.split() #Splitting lines into words at every whitespace

    #Defifning two lists: 1st- containing words that will be considered positive, 2nd- containing words that will be considered negative
        positive_words = ["good", "excellent", "satisfied", "happy","great","satisfactory","fair","best","impressive","exceptional","amazing","wonderful","fantastic","awesome","perfect","superb" ]
        negative_words = ["bad", "poor", "disappointed", "unhappy","unsatisfied","terrible","worst","pathetic","unsatisfactory","regret","awful",'ill',"worse"]

        for word in words: #iterating each and every word in a line
            mined_word = mine_word(word) #storing the cleaned word coming from the function into another variable
            if mined_word in positive_words: #checking if that word exists in the list of postive words
                sentiment = "Positive" #if yes then assigning it the category Positive
                pos_count += 1 #and incrementing the count
                pos_list.append(mined_word)
            
            elif mined_word in negative_words: #else checking if that word exists in the list of negative words
                sentiment = "Negative" #if yes then assigning it the category Negative
                neg_count += 1 #and incrementing the count
                neg_list.append(mined_word)

    return pos_count,pos_list,neg_count,neg_list #returning the values 

# Function defined to get insightd from the feedback
def per(pos_count,neg_count): 

    pos_per = float(pos_count/(pos_count+neg_count))*100 #calculated % of positive words out of total of +ve and -ve, typecasted to float
    neg_per = float(neg_count/(pos_count+neg_count))*100 #calculated % of negative words out of total of +ve and -ve, typecasted to float

    # Use of Conditional Statements checking % and provide recommendation for purchasing 
    if pos_per == neg_per:
        recommendation = 'Your choice whether to purchase as neutral reviews!!'
    elif pos_per >=70:
        recommendation = 'Highly recommended to purchase!!'
    elif pos_per > 50 and pos_per < 70:
        recommendation = 'Just Recommended to purchase!!'
    else:
        recommendation = 'Not recommended to purchase!!'

    return pos_per,neg_per,recommendation #returning the values
    
```

```python
  # DEFINING A FUNCTION FOR CALLING OF ALL OTHER FUNCTIONS AND DISPLAYING NECESSARY DETAILS

def call(feedback_data):
   
    pos_count, pos_list, neg_count, neg_list = feedback_analysis(feedback_data)

    print('Feedback Analysis: \n')
    print('No. of Positive Comments:',pos_count)
    print('List of positive words in the file:',pos_list)
    print('No. of Negative Comments:',neg_count)
    print('List of negative words in the file:',neg_list,'\n')

    pos_per, neg_per,recommendation = per(pos_count,neg_count)

    print('Insights \n')
    print('Percentage of positive words:',pos_per,'%')
    print('Percentage of negative words:',neg_per,'%')
    print('Status of recommendation:',recommendation)
```
